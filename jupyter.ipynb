There are two basic modes: 
* edit mode: press enter
* command mode: press esc. 

Edit mode allows you to edit the content of cells. Command mode allows you to manipulate cells using vim-like keybindings. 

The two most important kinds of cells are: 
* Markdown cells: press `m` in command mode
* Code cells: press `y` in command mode.
