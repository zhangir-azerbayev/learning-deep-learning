# String Formatting and Operations

```python
#C-style string formatting 
name = "Sam"
print("Hello, %s" % name)
```

Here are the basic argument specifiers
* %s - string
* %d - integers
* %f - floating point numbers
* %<no. digits>f - floating point number with fixed amount of digits past the dot

```python
# Tuples for multiple arguments
name = "Mike"
name2 = "Ike"
age = 27
print("%s is %d years old." % (name, age))
print("%s and %s are %d years old" % (name, name2, age))
```

```python
#Length
astring = "Hello world!"
print(len(astring))

# .index prints the index of the first occurence of a string
print(astring.index("wor"))

#slicing
print(astring[3:7])

#start:stop:step
print(astring[3:9:2])

# reverse a string
print(astring[::-1])

# convert to upper and lower case
print(astring.upper())
print(astring.lower())

#splits is cool
print(astring.split("ll"))
```

# Conditions and Loops

```python
# Equality
x = 2
print(x == 2)
print(x == 3)
print(x < 3)

# and, or
name = "Rick"
age = 55

if name == "Rick" and age == 55:
    print("your name is John and you are 55!")

if name == "Rick" or name == "John":
    print("your name is John or Rick")
    
# 'in' operator is used to check whether an object is in an iterable
names = ["Rick", "John"]

if name in names:
    print("cowabunga!")
    
# not quite clear on the difference between "is" and "=="

# There is a "not" operator

print(not True)
    
```

```python
primes = [2, 3, 5, 7]

for prime in primes:
    print(prime)

for x in range(3, 8, 2):
    print(x)

count = 0 
while count < 5: 
    print(count)
    count += 1

count = 0
while True:
    print(count)
    count += 1
    if count >= 5:
        break

# Prints out only odd numbers - 1,3,5,7,9
for x in range(10):
    # Check if x is even
    if x % 2 == 0:
        continue
    print(x)
    
#Python lets you use "else" at the end of a loop.
#If you 'break', the else part isn't executed. 
```

# Functions

```python
# As someone who learned functional languages first, this entire thing
# makes me super uncomfortable. 
# Define functions like this
def my_function():
    print("Hello from my function!")
    
# Functions can receive arguments
def func_with_args(name, greeting):
    print("Hello %s, from my function! I wish you %s" % (name, greeting))

# Functions can return values 
def sum_two_numbers(a, b):
    return a + b

# Call functions
func_with_args("John", "0 errors the first time you compile")
x = sum_two_numbers(1, 2)
print(x)
```

# Classes and Objects

```python
# Objects are a bag of variables and functions
# Classes are templates for creating objects
# You should take the name "object" quite literally. 

# I'll explain the "self" parameter later. 
class MyClass:
    variable = "blah"
    
    def function(self):
        print("This is a message inside the class")

myobjectx = MyClass()

print(myobjectx.variable)

# Object is a bag of variables 
myobjecty = MyClass()
myobjecty.variable = "hello"
print(myobjecty.variable)

#Access object functions
myobjectx.function()
```
