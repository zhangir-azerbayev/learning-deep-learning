# String Formatting and Operations

```python
#C-style string formatting 
name = "Sam"
print("Hello, %s" % name)
```

Here are the basic argument specifiers
* %s - string
* %d - integers
* %f - floating point numbers
* %<no. digits>f - floating point number with fixed amount of digits past the dot

```python
# Tuples for multiple arguments
name = "Mike"
name2 = "Ike"
age = 27
print("%s is %d years old." % (name, age))
print("%s and %s are %d years old" % (name, name2, age))
```

```python
#Length
astring = "Hello world!"
print(len(astring))

# .index prints the index of the first occurence of a string
print(astring.index("wor"))

#slicing
print(astring[3:7])

#start:stop:step
print(astring[3:9:2])

# reverse a string
print(astring[::-1])

# convert to upper and lower case
print(astring.upper())
print(astring.lower())

#splits is cool
print(astring.split("ll"))
```
